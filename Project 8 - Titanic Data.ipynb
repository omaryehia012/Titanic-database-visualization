```python
import seaborn as sns
%matplotlib inline

sns.set(rc={'figure.figsize': (9, 9)}, font_scale=1.2, style='whitegrid', palette='bright')
```

```python
df = sns.load_dataset('titanic')
```

```python
df.head()
```

```python
df.info()
```

```python
sns.distplot(df['age'], kde=False, bins=30, color='m')
```

```python
sns.kdeplot(df['age'], shade=True, color='m')
```

```python
sns.distplot(df['fare'], kde=False, bins=30, color='y')
```

```python
sns.kdeplot(df['fare'], shade=True, color='y')
```

```python
sns.jointplot(data=df, x='age', y='fare', kind='scatter', color='c')
```

```python
sns.jointplot(data=df[df['fare'] <= 100], x='age', y='fare', kind='scatter', color='c')
```
